#### Make a copy of the dataset

```python
df = google_apps.copy()
```

```python
df.shape
```

#### Drop NULL values

```python
df=df.dropna()
df.shape
```

#### Drop duplicated values

```python
df=df.drop_duplicates()
df.shape
```

#### Print first 10 values

```python
df.head(10)
```

#### Convert Date Time format

```python
df['Last_Updated'] = pd.to_datetime(df['Last Updated'])
df['day_updated'] = df['Last_Updated'].dt.day
df['year_updated'] = df['Last_Updated'].dt.year
df['month_updated']=df['Last_Updated'].dt.month
df['year_updated'].astype(int);
df['day_updated'].astype(int);
df['ContentRating']=df['Content Rating']
```

# Data Visualization

### Type: Free and paid apps

```python
sns.countplot(google_apps['Type'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Type')
```

### Rating of Google Play Store apps

```python
sns.countplot(google_apps['Rating'])
sns.countplot(google_apps['Rating']).set_xticklabels(sns.countplot(google_apps['Rating']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('Rating')
```

### Relation between Type and Rating

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Rating',hue='Type',data=google_apps)
plt.title('Relation between Type and Rating')
plt.show()
```

### Pie-chart for the Type: Free and Paid

```python
labels = ['Free', 'Paid', '0']
size = google_apps['Type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

### Pie-chart for Rating

```python
google_apps['Rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```
